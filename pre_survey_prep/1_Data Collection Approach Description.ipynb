# 1 Data Collection Approach Description

**Author:** Kristina Aleksandrovna Pedersen  


## Overview

This document describes how the political and entertainment TikTok video datasets were collected and prepared for the study. It focuses on the collection workflow, sample construction, and the final preprocessing steps applied before the material was inserted into the simulator.


## 1. Political Video Collection

We first compiled the TikTok profile URLs of all political user handles identified by the research assistants. We then used a scraper to automatically scroll through each profile page and collect both general user metadata and the URLs of all videos posted by each account.

After collecting the full set of video URLs from the initial user list, we scraped metadata for each video. This step excludes videos that required a TikTok account to view, such as age-restricted content, as well as videos that had been removed either by the platform or by the user between the two scraping steps. To reduce the risk of missing videos because of technical failures, we revisited the missing video URLs in two additional rounds.

The collected metadata included:

- video description
- timestamp
- duration
- posting location (not used in the analysis)
- TikTok's own content classification labels (not used in the analysis)
- user profile image

We then parsed the metadata and restricted the sample according to the predefined criteria described in Sections A and B of the Supplementary Materials. After this restriction step, we attempted to scrape the video files themselves. Not all videos in the restricted sample could be retrieved at this stage.


## 2. Entertainment Video Collection

We collected entertainment videos to increase the external validity of the TikTok-like social media simulator used in the experiment. Because the data collection took place outside Germany, we used six different German VPN locations while automatically scrolling the TikTok For You page. To keep recommendations as close to random as possible, the script hovered on videos for random intervals before continuing to the next video. Each scrolling session lasted 30 minutes.

As with the political sample, we then collected metadata for each video and the adjacent user profile. Using this metadata, we restricted the sample based on our predefined cutoffs.

Once we obtained the entertainment corpus, which contained 2,793 unique videos after removing duplicates, a sample was drawn and manually validated by research assistants. They were specifically instructed to flag videos containing political content and/or sensitive or offensive content. These videos were excluded from the dataset.

After restricting the sample to non-political and non-sensitive videos, we proceeded to scrape the video files themselves.


## 3. Difference Between the Two Samples

There is an important difference between the political and entertainment video samples. For the political sample, we began with the full population of interest: all videos posted by the political profiles identified by the research assistants. We then restricted this population based on the chosen cutoffs and, in some cases, video availability.

The entertainment sample is different. Rather than starting from a fixed population, it represents a sample of videos shown to a random non-user across different locations in Germany during December 2024 and January 2025.


## 4. Final Preprocessing Before Simulator Population

Before the simulator was populated with videos and metadata, all user profile images were rescaled to 200 x 200 pixels. If a user image was no longer available at the time of the final scrape, we used a blank placeholder image corresponding to TikTok profiles without a chosen profile photo.

In addition, all video encodings were converted to H.264 rather than H.265 before being inserted into the simulator.
